Add 1 to every age

```python
emp_age = np.array([22,25,28,31,35])
emp_age = emp_age + 1
print(emp_age)
```

10% hike on current salary

```python
salary = np.array([5,7,9,11])
new_salary = salary + (salary * 0.1)
print(new_salary)
```

shape , size , dtype

```python
data = np.array([100, 200, 300, 400, 500])
print(data.size)
print(data.shape)
print(data.dtype)
```

```python
emp = np.array([
    [22,5],
    [25,7],
    [30,10],
    [35,15]
])
print(emp.size)
print(emp.shape)
print(emp.dtype)
print(emp.ndim)
```

```python
print(emp[0,:])
print(emp[-1,:])
print(emp[:,0])
print(emp[:,-1])
print(emp[:2,:])
print(emp[-2:,:])
```

Concept 1 — Comparison Operators

```python
salary = np.array([5,7,10,15])

print(salary >= 10) #result - Boolean Mask
```

Concept 2 — Boolean Indexing

```python
salary = np.array([5,7,10,15])

print(salary[salary >= 10])
```

```python
emp = np.array([
    [22,5],
    [25,7],
    [30,10],
    [35,15]
])

salary = emp[:,1] #extract salary
print(salary)
print(salary[salary >= 10]) #find emps earning at least Rs. 10 LPA
```

aggregation functions

```python
print(salary.sum()) #Aggregation: sum
print(salary.mean()) #Aggregation: avg
print(salary.min()) #Aggregation: min
print(salary.max()) #Aggregation: max
print(salary.std()) #Aggregation: std
```

math operations

```python
salary = salary * 1.10 #increase salary by 10%
print(salary)
```

```python
salary = salary + 2 #bonus
print(salary)
```

```python
salary = salary - 1 #tax deduction
print(salary)
```

```python
age = emp[:,0]
salary = emp[:,1]

print((age > 25) & (salary > 8))
```

```python
print(emp[(age > 25) & (salary > 8)])
```

```python
salary = np.array([4,6,8,10,12])
print(salary[(salary > 8) & (salary < 10)])
```

```python
emp = np.array([
    [22,5],
    [25,7],
    [30,10],
    [35,15]
])
print(emp[emp[:,0] > 25])
print(emp[emp[:,1] > 7])
print(emp[(emp[:,0] > 25) & (emp[:,1] > 7)])
```

```python
marks = np.array([55,72,81,95,48,67])
print(marks.mean())
print(marks.max())
print(marks.min())
print(marks[marks > marks.mean()])
```
